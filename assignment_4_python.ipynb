```python
#Write a program to open the file romeo.txt and read it line by line. For each line, split the line into a list of words using the split function. For each word, check to see if the word is already in a list. If the word is not in the list, add it to the list. When the program completes, sort and print the resulting words in alphabetical

name = input("Enter file name: ")
file = open(name)
checklist= list()
for line in file:
    line=line.rstrip()
    word=line.split()
    i=0
    for z in word:
        if z not in checklist:
            checklist.append(z)
    else:
        continue
checklist.sort()
print (checklist)
```
